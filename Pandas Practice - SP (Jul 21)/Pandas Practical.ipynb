## __Section 7: Reshaping Data__

### __A. Drop Columns / Rows from a DataFrame__

### __A. Task 1__

```python
df = pd.read_csv("Pandas Practical Datasets/graduate-employment-survey-ntu-nus-sit-smu-sutd-utf8.csv", header=0,
                sep=",", na_values="na") # Include the paramter "na_values" otherwise the numeric columns will be object instead of float

print("DataFrame.info:\n", df.info(), "\n")
print("DataFrame.dtypes:\n", df.dtypes, "\n")
```

```python
print("Shape of df in its original from = {}".format(df.shape))

# Operation 1

dropped_cols = ["school","employment_rate_overall","employment_rate_permanent",
        "basic_monthly_mean","basic_monthly_median","gross_mthly_25_percentile",
               "gross_mthly_75_percentile"]
df = df.drop(labels=dropped_cols,axis=1)
print("Shape of df after dropping unwanted columns = {}".format(df.shape))

# Operation 2

dropped_uni_list = ["Nanyang Technological University","Singapore Institute of Technology", "Singapore University of Technology and Design"]
df = df.drop(labels=list(df[df["university"].isin(dropped_uni_list)].index))
print("Shape of df after dropping unwanted rows (university) = {}".format(df.shape))

# Operation 3

dropped_year = [2013,2014]
# df = df[~df["year"].isin(dropped_year)] # I prefer this since it's more elegant
df = df.drop(df[df["year"].isin(dropped_year)].index)
print("Shape of df after dropping unwanted rows (year) = {}".format(df.shape))

# Operation 5

dropped_gross_monthly_median = df["gross_monthly_median"]<4000
df = df[~dropped_gross_monthly_median]
print("Shape of df after dropping unwanted rows (gross_monthly_median) = {}".format(df.shape))

print("***Final data after all the drops ***");print(df)
```

### __D: Gather Columns into Rows (Melt)__

### __D. Task 1__

```python
mydf = pd.read_csv('Pandas Practical Datasets/irascollectionbytaxtypemelt.csv',sep=',',header=0)

print('***** Before MELT *****\n\n{}'.format(mydf.head(5)))
print()

# Unpivot data from columns to rows via melt

# First we find out what columns there are with: mydf.columns, then we extract them into a list below
financial_year = ['2002', '2003', '2004', '2005', '2006', '2007', '2008',
                  '2009', '2010', '2011', '2012', '2013', '2014', '2015']
mydf = pd.melt(mydf, id_vars=['tax_type'], value_vars=financial_year)

# Rename the columns then reindex the columns
mydf.rename(columns={'variable':'financial_year', 'value':'tax_collected'})
new_index = ['financial_year', 'tax_type', 'tax_collected']

print('***** After MELT *****\n\n{}'.format(mydf.head(5)))
```

## __Section 9: Combine Datasets__

### __A. Merging Datasets__

```python
df_m = pd.read_csv('Pandas Practical Datasets/merge_students-male.csv', sep=',', header=0, index_col=0)
df_f = pd.read_csv('Pandas Practical Datasets/merge_students-female.csv', sep=',', header=0, index_col=0)
df_test = pd.read_csv('Pandas Practical Datasets/merge_testscores.csv', sep=',', header=0, index_col=0)

df_students = pd.concat([df_m,df_f]) # Concatenate the students' datasets together

df_merged = df_students.merge(df_test, how='inner', on='studentid')
df_merged_sorted = df_merged.sort_values(by='test score', ascending=False)

print(f'Total students: {len(df_merged)}\n\nTest scores in descending order\n{df_merged_sorted}')
```
